```python
df=pd.read_csv("cleanedDataset.csv")
```

```python
df.head()
```

```python
source_count=df.groupby(["site_url","label"]).size().unstack(fill_value=0)
```

```python
source_count["Percentage Real(%)"]=(source_count["Real"]/(source_count["Real"]+source_count["Fake"]))* 100
```

```python
source_count["Percentage Fake(%)"]=(source_count["Fake"]/(source_count["Real"]+source_count["Fake"]))* 100
```

```python
sorted_sources=source_count.sort_values(by="Percentage Real(%)",ascending=False)
```

```python
print("Top 10 most credible News Sources:")
for source,rows in sorted_sources.head(10).iterrows():
  print(f" News {source},fake news={rows['Percentage Fake(%)']:.1f}%")

print("Top 10 least credible News Sources:")
for source,rows in sorted_sources.tail(10).iterrows():
  print(f"News {source},fake news={rows['Percentage Fake(%)']:.1f}%")
```
